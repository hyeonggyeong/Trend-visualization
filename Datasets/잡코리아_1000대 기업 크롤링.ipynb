```python
# 페이지에 있는 공고들의 기업들이름을  리스트에 수집하는 함수
def get_company_name(list_name):
    elems = driver.find_element_by_css_selector('#container > section > div.readSumWrap.clear > article > div.sumTit > h3 > div > span').text
    list_name.append(elems)
    time.sleep(1)
    
# 페이지에 있는 기업 리스트 중 하나를 골라 들어간 후 관심을 보인 사람들 수를 뽑는 함수
# 리스트를 입력 받아서 관심 사람 수를 추가.
def get_attention(list_1):
    try: 
        attention =driver.find_element_by_css_selector('#tab01 > article.artReadReason > ul > li.icn_1 > strong').text
        list_1.append(attention)
        time.sleep(1)
        
    except:
        list_1.append('NaN')
        
# 페이지에 들어가서 회사 설립년도수를 뽑는 함수
def get_hist(list_2):
    try:
        hist =driver.find_element_by_css_selector('#container > section > div.readSumWrap.clear > article > div.tbRow.clear > div.tbCol.tbCoInfo > dl > dd:nth-child(6) > text > span:nth-child(2)').text
        list_2.append(int(hist))
        time.sleep(1)
        
    except:
        list_1.append('NaN')
        
# 페이지에 들어가서 회사 규모를 뽑는 함수
def get_scale(list_3):
    scale = driver.find_element_by_css_selector('#container > section > div.readSumWrap.clear > article > div.tbRow.clear > div.tbCol.tbCoInfo > dl > dd:nth-child(8)').text
    list_3.append(scale)
    time.sleep(1)
    
# 입사복지정보들을 빼는 함수 
def get_welfare(list_4):
    try:
        wel_table = driver.find_element_by_css_selector('#secReadWork > article > div.location > div > dl > dd').text
        wel_num = len(attention.split(','))-len(attention.split('\n'))         # 적혀있는 복지 혜택이 몇개 있는가 산출 
        time.sleep(1)
        
    except:
        list_4.append(0)
        
# 입사 지원자 현황을 빼는 함수 
# 결과 형태는 (지원자,모집인원)
def get_apply(list_5):
    try:
        apply = driver.find_element_by_css_selector('#detailArea > section.secReadStatistic > article > div.metricsContainer > div.metrics.metricsCount > div.value > em').text
        want = driver.find_element_by_css_selector('#detailArea > section.secReadStatistic > article > div.metricsContainer > div.metrics.metricsRate > div.value > em').text
        total = tuple((int(apply),int(want)))
        list_5.append(total)
        time.sleep(1)
        
    except:
        list_5.append('NaN')
```

## 페이지로부터 hrep 추출하는 함수

```python
# 페이지를 입력받아서 그 페이지의 기업페이지 href를 출력받음
# 밑에 page_to를 만들기 위한 함수 
def get_data_page(page_num):
    driver = webdriver.Chrome()
    driver.get('https://www.jobkorea.co.kr/Starter/?JoinPossible_Stat=0&schOrderBy=0&LinkGubun=1&LinkNo=0&schType=0&schGid=0&Page={0}'.format(page_num))
    href_list =[]
    time.sleep(1)
    try:
        for i in range(1,41):
            select = '#devStarterForm > div.filterListArea > ul > li:nth-child({0}) > div.info > div.tit > a'.format(str(i))
            href =driver.find_element_by_css_selector(select).get_attribute('href')
            href_list.append(href)
            time.sleep(1)
        driver.close()
    except:
        driver.close()
                
    return href_list

# page_to(페이지수) 를 입력하면 입력한 페이지 까지 있는 모든 회사페이지들의 href를 구할 수 있다.
def page_to(page_num):
    total_href =[]
    
    for i in range(1,page_num+1):
        total_href=total_href+get_data_page(i)
        
    return total_href
                       
```

## href 리스트 들을 입력받아 회사 페이지 내에 정보들을 추출하는 함수 

```python
def get_data(href_list,list_1,list_2,list_3,list_4,list_5,list_name):
    for href in href_list:
        driver.get(href)
        get_attention(list_1)
        get_hist(list_2)
        get_scale(list_3)
        get_welfare(list_4)
        get_apply(list_5)
        get_company_name(list_name)
        time.sleep(1.5)    
```
